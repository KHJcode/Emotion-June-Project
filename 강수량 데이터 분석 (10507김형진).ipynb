## 문제 : 연도를 입력받고, 해당 연도의 1월과 8월 일자별 강수량을 [그래프 조건]에 맞춰 선 그래프로 시각화하시오. 시각화 한 후에는 마지막에 데이터 분석결과를 적어보자.
### [그래프 조건]
0. 그래프 모양 : 선 그래프
1. 선 그래프 제목 : ~년도 강수량 (*한글 깨지지 않도록 처리하기)
2. 선 그래프 각각에 범례 추가 (1월, 8월)
3. x축 레이블 추가 (일자)
4. y축 레이블 추가 (강수량)
5. 1월 그래프는 파란색, 8월 그래프는 빨간색으로 설정
6. 제출 시 연도에 2020을 입력하여 출력된 상태로 저장하여 제출한다.

### [절차]
1. rain.csv 파일과 현재 작성하고 있는 주피터 노트북 파일을 자신의 바탕화면(Desktop)에 저장한다. (파일 이름 정확한지 확인, 파일 깨지지 않았는지 확인)
2. rain.csv 파일을 주피터 노트북에서 읽을 수 있도록 처리한다.
3. 1월, 8월 데이터를 저장할 빈 리스트 2개를 생성하고, 연도를 입력받아 변수에 저장한다.
4. 데이터를 읽어오면서 데이터 중에서 강수량에 해당하는 부분이 공백이면 0으로 대체시킨다.
5. 데이터를 읽어오면서 데이터 중에서 날짜에 해당하는 부분을 연도, 월, 일로 분리했을 때, 연도가 입력받은 연도와 같고 월이 '8'이라면, 8월 데이터 리스트에 강수량을 실수형으로 저장한다. 1월 강수량 데이터도 동일하게 저장한다.
6. 위 [그래프 조건]에 맞춰서 선 그래프로 시각화한다.
7. 2018년도와 2020년도의 그래프를 분석한 결과를 자신의 생각대로 적어본다.
8. 최종 출력은 2020을 출력한 상태로 저장한다.

```python
import csv
import matplotlib.pyplot as plt

f = open('rain.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')

jan = []
aug = []
year = input('연도를 입력하세요: ')

for row in data:
    if row[-1] == '':
        row[-1] = 0
    if row[0].split('.')[0] == year:
        if row[0].split('.')[1] == '8':
            aug.append(float(row[-1]))
        if row[0].split('.')[1] == '1':
            jan.append(float(row[-1]))

print(jan)
print(aug)
            
plt.rc('font', family='Malgun Gothic')
plt.title(year + '년도 강수량')
plt.plot(aug, label='8월', color='r')
plt.plot(jan, label='1월', color='b')
plt.xlabel('일자')
plt.ylabel('강수량')
plt.legend()
plt.show()

f.close()
```
